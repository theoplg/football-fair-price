```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Visualization 

```python
# 1. Charger tes données propres
df = pd.read_csv('../data/processed/transfers_ready_for_model.csv')

# 2. Voir un aperçu
display(df.head())

# 3. GRAPHIQUE 1 : Le Prix selon le Poste
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Position_Encoded', y='Prix')
plt.title("Distribution des prix par poste (1=Gardien, 4=Attaquant)")
plt.show()

# 4. GRAPHIQUE 2 : Corrélation Age / Prix
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Prix', hue='Position_Encoded')
plt.title("Prix selon l'âge")
plt.show()
```
